### Average height and weight overall

```python
print(f"Average Height: {df['height'].mean():.2f}")
print(f"Average Weight: {df['weight'].mean():.2f}")

sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['height'], kde=True, ax=ax[0], color='Pink').set_title('Height Distribution')
sns.histplot(df['weight'], kde=True, ax=ax[1], color='navy').set_title('Weight Distribution')
plt.show()
```

### relationship between age and height/weight

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.set_theme(style="whitegrid")
sns.regplot(x='age', y='height', data=df, ax=ax[0]).set_title('Age vs Height')
sns.regplot(x='age', y='weight', data=df, ax=ax[1], color='orange').set_title('Age vs Weight')
plt.show()
```
